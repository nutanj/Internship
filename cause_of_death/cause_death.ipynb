```python
def outlierRemover(col):
    q1,q2= col.quantile([0.25,0.75])
    IQR= q2-q1
    lowerRange= q1-(1.5*IQR)
    upperRange= q2+(1.5*IQR)
    return lowerRange,upperRange
```

```python
lowScore, highScore= outlierRemover(df['Meningitis'])
```

```python
lowScore, highScore
```

```python
df['Meningitis']= np.where(df['Meningitis']>highScore, highScore, df['Meningitis'])
df['Meningitis']= np.where(df['Meningitis']<lowScore, lowScore, df['Meningitis'])
```
